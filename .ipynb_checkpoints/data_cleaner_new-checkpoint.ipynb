# Check all avaliable Tickers

```python
import gc
import os, glob
import pandas as pd
```

```python
path = r'All Data\DiscountOptionsData\DataByDate'
aaall_tickers = set()
days = [ str(i).rjust(2, '0') for i in range(1,30)]
yrs = [ str(i) for i in range(2005,2023)]
months = [ str(i).rjust(2, '0') for i in range(1,13)]

mo_counter = 0
yr_counter = 0

for filename in glob.glob(os.path.join(path, '*.csv')):
    if mo_counter >= len(months):
        mo_counter = 0 
        yr_counter += 1
        
    if yr_counter >= len(yrs):
        break
    check_date = yrs[yr_counter] + months[mo_counter]
    if filename[46:52] == check_date:
        if filename[52:54] in days:
            _df = pd.read_csv(filename, low_memory = False)
            ticks = list(_df.Symbol.unique())
            aaall_tickers.update(ticks)
            mo_counter += 1
            print(filename)
```

# Parse files by symbol

```python
import gc
import os, glob
import pandas as pd
import time
path = r'DiscountOptionsData\2010_2022'
# all_tickers = pd.read_csv('All_Symbols.csv')
# tickers = set(all_tickers['0'])
yrs = [ str(i) for i in range(2020,2023)]
months = [ str(i).rjust(2, '0') for i in range(1,13)]
# tickers = ['SPY','AAPL','TSLA','NFLX']

for yr in yrs:
    time0 = time.time()
    for mo in months:
        time2 = time.time()
        mo_df = pd.DataFrame()
        for filename in glob.glob(os.path.join(path, '*.csv')):
#             print(filename)
        #     with open(os.path.join(os.getcwd(), filename), 'r') as f: # open in readonly mode
            dt_id = filename[36:42]
 
            if (dt_id == (yr + mo)):
                _df = pd.read_csv(filename,  dtype={'ImpliedVolatility': 'object', 
                                                    'Rho':'object','vega':'object','Theta':'object'}
                                  ,usecols = [i for i in range(1,20)])
                mo_df = pd.concat([mo_df,_df], ignore_index = True)
                del _df

        mo_df.rename({'DataDate':'QuoteDate'},axis=1,inplace = True)
        if len(mo_df) == 0:
            print('Missing Data', yr, mo)
            continue
        tickers = set(mo_df.Symbol)
        for tick in tickers:
            if(tick == 'ULTEE'):
                continue
            t_df = mo_df.loc[mo_df.Symbol == tick]
            if (len(t_df)):
                try:
                    t_df.to_csv('DataBySymbol\\'+ tick+ '\\' + tick+'_'+yr+mo+'.csv',index = False)
                except:
                    try:
                        os.mkdir('DataBySymbol\\' + tick) 
                        t_df.to_csv('DataBySymbol\\'+ tick+ '\\' + tick+'_'+yr+mo+'.csv',index = False)
                        print('Make Folder:',tick)
                    except:
                        print('Cannot Make Folder:',tick)
                        pass
            del t_df
        time3 = time.time()
        print(yr, mo, 'Done',time3 - time2,'seconds')
        del mo_df
        gc.collect()
   
    time1 = time.time()
    print(yr,'Done',time1 - time0,'seconds')
```
